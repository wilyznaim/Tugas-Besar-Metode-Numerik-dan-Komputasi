# Soal 1
## Hitung dengan metode Newton-Raphson

$ K_{p} D^3 - K_{A}(H+D)^3 = 0 $

dengan :

$ K_{A} = tan^2(45°-\Phi/2) $

$ K_{p} = tan^2(45°+\Phi/2) $

### Ditanya : kedalaman pemancang dinding turap baja (D)
### Penyelesaian :

$ H = 10 + (X + Z) $

$ \Phi = 2Y,ZX° $

```python
import math
import sympy as sp
from prettytable import PrettyTable
```

```python
x = sp.Symbol('x')

# Acakan X,Y,Z
print('[!] Masukkan nilai acakan XYZ')
acakan_X = int(input('X : ')) #Input nilai acakan X
acakan_Y = int(input('Y : ')) #Input nilai acakan Y
acakan_Z = int(input('Z : ')) #Input nilai acakan Z

print('\n[+] Diketahui')
H = 10 + (acakan_X + acakan_Z) #Menghitung nilai H
sudut = float(f'2{acakan_Y}.{acakan_Z}{acakan_X}') #Menghitung sudut dan merubah menjadi float
print(f'H : {H}')
print(f'Φ : {sudut}')

Ka = math.tan(math.radians(45 - sudut/2))**2 #Menghitung Ka
Kp = math.tan(math.radians(45 + sudut/2))**2 #Menghitung Kp
persamaan = Kp * (x**3) - Ka * (H + x) ** 3 #Deklarasi persamaan pada soal
turunan = persamaan.diff(x, 1) #Turunan pertama dari persamaan

print('\nKa :', Ka)
print('Kp :', Kp)

# Input
print('\n[!] Masukkan nilai xi awal')
x1 = float(input('xi : ')) #Input x1 sebagai nilai awal
error_toleransi = float(input('Toleransi (gunakan . sebagai koma) : ')) #Input toleransi maksimal yang diperbolehkan

# Nilai awal
i = 0 #Nilai awal jumlah iterasi
e = 1 #Nilai awal error toleransi

# Perhitungan
print(f'\n[+] Perhitungan Newton-Raphson dengan nilai awal {x1}')
table = PrettyTable(['Iterasi', 'xi', 'xi+1', 'f(xi)', 'f\'(xi)', 'f(xi+1)']) #Membuat head cli-table
oncemore = iter([True, False]) #Deklarasi variabel untuk menambah 1 kali iterasi setelah iterasi dibawah selesai
while e >= error_toleransi or next(oncemore):
  #Selama error lebih besar error_toleransi, iterasi tidak berhenti
  #Jika sudah tercapai e >= error_toleransi, akan dilakukan 1 kali iterasi lagi dari fungsi next()
  f1 = persamaan.subs(x, x1) #Subtitusi x1 (xi) pada persamaan
  f2 = turunan.subs(x, x1) #Subtitusi x1 (xi) pada turunan pertama persamaan
  x2 = x1 - (f1/f2) #Menghitung x2 (xi+1)
  f3 = persamaan.subs(x, x2) #Subtitusi x1 (xi+1) pada persamaan
  e = abs(x1 - x2) #Menghitung error
  
  i = i+1 #Menambah nilai iterasi ke-
  table.add_row([i, f'{float(x1):.4f}', f'{float(x2):.4f}', f'{float(f1):.4f}', f'{float(f2):.4f}', f'{float(f3):.4f}']) #Input hasil pada tabel
  x1 = x2 #Deklarasi xi (xi) menggunakan nilai x2 (xi+1)

print(table) #Membuat tabel

print('\n[+] KONTROL')
print(f'D : {x1}') #Memperoleh x1 dari hasil iterasi
print(f'Toleransi : {error_toleransi}')
print(f'{persamaan} < {error_toleransi}')
print(f'{persamaan.subs(x, x1)} < {error_toleransi} OK') if persamaan.subs(x, x1) <= error_toleransi else print(f'{persamaan.subs(x, x1)} > {error_toleransi} SALAH')
#Jika subtitusi x1 (D) pada persamaan awal < toleransi, maka terpenuhi
```
